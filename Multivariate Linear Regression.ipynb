Importing all libraries and reading the data


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('data.csv')
df.head()
```

Dropping all irrelevant colums

```python
df=df.drop(['MODELYEAR','MAKE','MODEL','VEHICLECLASS','ENGINESIZE','CYLINDERS','TRANSMISSION','FUELTYPE'], axis=1)
df
```

Adding all elements of the 2 features into a single numpy array, and the Y feature values in a different array

```python
X1=df['FUELCONSUMPTION_CITY'].values
X2=df['FUELCONSUMPTION_HWY'].values
Y=df['CO2EMISSIONS'].values
X=[]
for i in range(len(X1)):
  X.append([float(X1[i]),float(X2[i]),1])
X=np.array(X)
Y=np.array(Y)
```

Applying Multivariate Linear Regression of 2 variables

```python
w=np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
Y_pred=np.dot(X,w)
Y_pred
```

Checking accuracy of the model using R-squared value or coeffecient of determination

```python
rss=Y-Y_pred
tss=Y-Y.mean()
r=1-(rss.dot(rss)/tss.dot(tss))
r
```
